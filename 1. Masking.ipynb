```python
import os
from pathlib import Path
from rembg import remove
from PIL import Image
import tensorflow as tf
from io import BytesIO
```

```python
def process_and_save_image(image_path, save_path):
    try:
        # Load the image
        with open(image_path, 'rb') as i:
            input_image = i.read()

        # Remove background
        output_image = remove(input_image)

        # Open the output image as a PIL Image
        img = Image.open(BytesIO(output_image)).convert("RGB")

        # Resize the image to 590x445 pixels
        img = img.resize((590, 445), Image.Resampling.LANCZOS)

        # Save the image
        img.save(save_path)
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
```

```python
def process_directory(input_dir):
    input_dir = Path(input_dir)
    output_dir = input_dir.parent / (input_dir.name + " Masked")
    output_dir.mkdir(exist_ok=True)

    # Valid image extensions
    valid_extensions = {'.jpg', '.jpeg', '.png'}
    
    # Counter for serial numbering
    counter = 1
    
    for image_path in input_dir.glob('*.*'):
        if image_path.suffix.lower() in valid_extensions:
            # Generate a new filename with serial number and same extension
            new_filename = f"{counter:04d}{image_path.suffix.lower()}"
            save_path = output_dir / new_filename
            
            # Process and save the image
            process_and_save_image(image_path, save_path)
            
            # Increment the counter
            counter += 1
```

```python
# Example usage
input_folder = 'D:\OUR COLLECTED'
for sub_dir in Path(input_folder).iterdir():
    if sub_dir.is_dir():
        process_directory(sub_dir)
```
